def ConvWeightNorm(optimizer):
    layers = 0
    for param in optimizer.param_groups[0]['params']:
        if len(param.shape) == 4 and layers < 3:
            abs_sum = param.data.abs().sum(3,True).sum(2,True).sum(1,True)
            param.data.div_(abs_sum)
            layers+=1
